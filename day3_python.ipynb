```python
# neural activity simulation, normalization, and plotting

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Parameters
# -------------------------------
n_neurons = 5       # number of neurons
n_timepoints = 20   # number of time points

# -------------------------------
# 2️⃣ Simulate neural activity
# -------------------------------
# Random numbers between 0 and 1
neural_activity = np.random.rand(n_neurons, n_timepoints)

print("Simulated neural activity:")
print(neural_activity)

# -------------------------------
# 3️⃣ Normalize the activity
# -------------------------------
def normalize_neural_data(data, method="minmax"):
    """
    Normalize neural activity data.
    
    Parameters:
        data (2D array): shape (n_neurons, n_timepoints)
        method (str): "minmax" or "zscore"
    
    Returns:
        2D array: normalized data
    """
    data = np.array(data, dtype=float)
    
    if method == "minmax":
        norm_data = (data - data.min(axis=1, keepdims=True)) / \
                    (data.max(axis=1, keepdims=True) - data.min(axis=1, keepdims=True))
    elif method == "zscore":
        norm_data = (data - data.mean(axis=1, keepdims=True)) / data.std(axis=1, keepdims=True)
    else:
        raise ValueError("Method must be 'minmax' or 'zscore'")
    
    return norm_data

# Normalize with min-max
norm_minmax = normalize_neural_data(neural_activity, method="minmax")

# Normalize with z-score
norm_zscore = normalize_neural_data(neural_activity, method="zscore")

# -------------------------------
# 4️⃣ Plot the activity
# -------------------------------
plt.figure(figsize=(12, 5))

# Original activity
plt.subplot(1, 3, 1)
for neuron in neural_activity:
    plt.plot(neuron)
plt.title("Original activity")
plt.xlabel("Time")
plt.ylabel("Activity")

# Min-max normalized
plt.subplot(1, 3, 2)
for neuron in norm_minmax:
    plt.plot(neuron)
plt.title("Min-max normalized")
plt.xlabel("Time")

# Z-score normalized
plt.subplot(1, 3, 3)
for neuron in norm_zscore:
    plt.plot(neuron)
plt.title("Z-score normalized")
plt.xlabel("Time")

plt.tight_layout()
plt.show()
```

```python
# spiking activity simulation to normalize it

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Parameters
# -------------------------------
n_neurons = 5       # number of neurons
n_timepoints = 50   # number of time points
spike_prob = 0.2    # probability a neuron fires at each time point

# 2️⃣ Simulate spiking activity

# Generate spikes: 1 = spike, 0 = no spike
spiking_activity = np.random.rand(n_neurons, n_timepoints) < spike_prob 
# np.random.rand(n_neurons, n_timepoints) generates a matrix of random numbers between 0–1.
spiking_activity = spiking_activity.astype(int)
# < spike_prob converts it to boolean: True if the random number < 0.2 (i.e., neuron fires). .astype(int) converts True/False → 1/0.

# -------------------------------
# 3️⃣ Normalize activity
# -------------------------------
# Min-max normalization for each neuron independently
norm_spiking_minmax = (spiking_activity - spiking_activity.min(axis=1, keepdims=True)) / \
                      (spiking_activity.max(axis=1, keepdims=True) - spiking_activity.min(axis=1, keepdims=True))

# Z-score normalization for each neuron independently
norm_spiking_zscore = (spiking_activity - spiking_activity.mean(axis=1, keepdims=True)) / \
                      spiking_activity.std(axis=1, keepdims=True)

# -------------------------------
# 4️⃣ Plot original spiking activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(spiking_activity):
    plt.scatter(range(n_timepoints), neuron + i*1.5, marker='|', s=200)
plt.yticks(np.arange(0, n_neurons*1.5, 1.5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Original Spiking Neural Activity")
plt.show()

# -------------------------------
# 5️⃣ Plot min-max normalized activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(norm_spiking_minmax):
    plt.plot(neuron + i*1.5, marker='o', linestyle='-', markersize=4)
plt.yticks(np.arange(0, n_neurons*1.5, 1.5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Min-Max Normalized Neural Activity")
plt.show()

# -------------------------------
# 6️⃣ Plot z-score normalized activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(norm_spiking_zscore):
    plt.plot(neuron + i*1.5, marker='o', linestyle='-', markersize=4)
plt.yticks(np.arange(0, n_neurons*1.5, 1.5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Z-Score Normalized Neural Activity")
plt.show()
```

```python
# simulate smooth neural firing rates (not just spikes) and normalize them for comparison with this spiking activity
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Parameters
# -------------------------------
n_neurons = 5       # number of neurons
n_timepoints = 50   # number of time points

# -------------------------------
# 2️⃣ Simulate smooth firing rates
# -------------------------------
# Generate random baseline activity
np.random.seed(42)  # for reproducibility
smooth_activity = np.random.rand(n_neurons, n_timepoints) * 20  # firing rates in Hz

# Add some temporal structure (smooth variation)
time = np.arange(n_timepoints)
for i in range(n_neurons):
    smooth_activity[i] += np.sin(time / 5 + i) * 5  # sine modulation

# -------------------------------
# 3️⃣ Normalize activity
# -------------------------------
# Min-max normalization for each neuron independently
norm_smooth_minmax = (smooth_activity - smooth_activity.min(axis=1, keepdims=True)) / \
                     (smooth_activity.max(axis=1, keepdims=True) - smooth_activity.min(axis=1, keepdims=True))

# Z-score normalization for each neuron independently
norm_smooth_zscore = (smooth_activity - smooth_activity.mean(axis=1, keepdims=True)) / \
                     smooth_activity.std(axis=1, keepdims=True)

# -------------------------------
# 4️⃣ Plot original smooth activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(smooth_activity):
    plt.plot(neuron + i*25, marker='o')  # offset for clarity
plt.yticks(np.arange(0, n_neurons*25, 25), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Original Smooth Neural Activity (Firing Rates in Hz)")
plt.show()

# -------------------------------
# 5️⃣ Plot min-max normalized smooth activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(norm_smooth_minmax):
    plt.plot(neuron + i*1.5, marker='o')
plt.yticks(np.arange(0, n_neurons*1.5, 1.5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Min-Max Normalized Smooth Neural Activity")
plt.show()

# -------------------------------
# 6️⃣ Plot z-score normalized smooth activity
# -------------------------------
plt.figure(figsize=(10, 4))
for i, neuron in enumerate(norm_smooth_zscore):
    plt.plot(neuron + i*1.5, marker='o')
plt.yticks(np.arange(0, n_neurons*1.5, 1.5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.xlabel("Time")
plt.title("Z-Score Normalized Smooth Neural Activity")
plt.show()
```

```python
# combine spikes and smooth activity into one plot per neuron
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Parameters
# -------------------------------
n_neurons = 5
n_timepoints = 50
spike_prob = 0.2  # probability of a spike at each time point

np.random.seed(42)  # reproducibility

# -------------------------------
# 2️⃣ Simulate spikes (binary)
# -------------------------------
spikes = np.random.rand(n_neurons, n_timepoints) < spike_prob
spikes = spikes.astype(int)  # 1 if spike, 0 otherwise

# -------------------------------
# 3️⃣ Simulate smooth firing rates
# -------------------------------
smooth_activity = np.random.rand(n_neurons, n_timepoints) * 20
time = np.arange(n_timepoints)
for i in range(n_neurons):
    smooth_activity[i] += np.sin(time / 5 + i) * 5

# -------------------------------
# 4️⃣ Normalize smooth activity (z-score)
# -------------------------------
norm_smooth = (smooth_activity - smooth_activity.mean(axis=1, keepdims=True)) / \
              smooth_activity.std(axis=1, keepdims=True)

# -------------------------------
# 5️⃣ Plot combined
# -------------------------------
plt.figure(figsize=(12, 6))

for i in range(n_neurons):
    # Plot smooth activity
    plt.plot(norm_smooth[i] + i*5, label=f"Neuron {i+1} smooth")
    
    # Plot spikes as vertical markers
    spike_times = np.where(spikes[i] == 1)[0]
    plt.vlines(spike_times, i*5, i*5 + 1, color='r', linewidth=2)

plt.xlabel("Time")
plt.ylabel("Neuron / Normalized Firing Rate")
plt.title("Combined Spikes (red) and Smooth Neural Activity (lines)")
plt.yticks(np.arange(0, n_neurons*5, 5), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1️⃣ Parameters
# -------------------------------
n_neurons = 5
n_timepoints = 50
spike_prob = 0.2

np.random.seed(42)

# -------------------------------
# 2️⃣ Simulate spikes
# -------------------------------
spikes = (np.random.rand(n_neurons, n_timepoints) < spike_prob).astype(int)

# -------------------------------
# 3️⃣ Simulate smooth firing rates
# -------------------------------
smooth_activity = np.random.rand(n_neurons, n_timepoints) * 20
time = np.arange(n_timepoints)
for i in range(n_neurons):
    smooth_activity[i] += np.sin(time / 5 + i) * 5

# -------------------------------
# 4️⃣ Normalize
# -------------------------------
# Z-score normalization
zscore_norm = (smooth_activity - smooth_activity.mean(axis=1, keepdims=True)) / \
              smooth_activity.std(axis=1, keepdims=True)

# Min-max normalization
minmax_norm = (smooth_activity - smooth_activity.min(axis=1, keepdims=True)) / \
              (smooth_activity.max(axis=1, keepdims=True) - smooth_activity.min(axis=1, keepdims=True))

# -------------------------------
# 5️⃣ Plot
# -------------------------------
plt.figure(figsize=(14, 6))

for i in range(n_neurons):
    # Z-score normalized activity (solid line)
    plt.plot(zscore_norm[i] + i*3, label=f"Neuron {i+1} Z-score")
    
    # Min-max normalized activity (dashed line)
    plt.plot(minmax_norm[i] + i*3, '--', label=f"Neuron {i+1} Min-Max")
    
    # Spikes as red vertical lines
    spike_times = np.where(spikes[i] == 1)[0]
    plt.vlines(spike_times, i*3, i*3 + 0.5, color='r', linewidth=2)

plt.xlabel("Time")
plt.ylabel("Neuron / Normalized Firing Rate")
plt.title("Neural Activity: Z-score vs Min-Max Normalization + Spikes")
plt.yticks(np.arange(0, n_neurons*3, 3), [f"Neuron {i+1}" for i in range(n_neurons)])
plt.legend(loc='upper right', fontsize=8)
plt.show()
```
